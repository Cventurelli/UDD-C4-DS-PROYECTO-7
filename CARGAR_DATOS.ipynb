Procedimiento para descargar y descomprimir el conjunto de datos "Chest X-ray Pneumonia" de Kaggle:

Descarga del conjunto de datos:

Para descargar el conjunto de datos "Chest X-ray Pneumonia" de Kaggle, se utilizó el siguiente comando:

!kaggle datasets download -d paultimothymooney/chest-xray-pneumonia
Este comando descarga el conjunto de datos en la carpeta actual.

Descompresión del archivo zip:

Una vez descargado el conjunto de datos, se descomprimió el archivo zip utilizando el siguiente comando:

!unzip chest-xray-pneumonia.zip
Este comando extrae el contenido del archivo zip en la carpeta actual.
